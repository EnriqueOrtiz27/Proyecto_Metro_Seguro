# El propósito de estas funciones es simplemente limpiar la base de datos del gobierno para después usarla en las siguientes etapas del proceso. Esta base de datos es la del gobierno (con los 800k+ delitos).

```python
data = pd.read_csv('carpetas-de-investigacion-pgj-cdmx.csv', sep=';')
```

```python
def exe(data):
    """En esta función voy a agrupar los métodos que usé para limpiar la base de datos."""
    data = data.drop(columns=['fiscalía', 'agencia', 'unidad_investigacion', 'fecha_inicio', 'mes_inicio', 
                          'ao_inicio', 'calle_hechos2', 'Geopoint'])
    data = data.rename(columns={'index':'id'})
    data = data.dropna()
    data = data.reset_index()
    data.fecha_hechos = pd.to_datetime(data.fecha_hechos)
    
    return data

data = exe(data)
```

```python
res = []
for i in range(len(data)):
   
    geopunto = {}
    if data['longitud'][i] and data['latitud'][i]:
        geopunto ={'type':'Point',
                       'coordinates':[
                           data['longitud'][i],
                           data['latitud'][i],
                       ]}
        res.append(geopunto)

data['geo_punto'] = res
```
